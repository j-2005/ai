  * 높임말 번역 기능, 오프라인 번역 기능, 웹 페이지 번역 기능 지원

2. 파파고에 사용된 기술<br>
  * **인공신경망 기계번역(NMT) 기술**
    * *Neural Machine Translation(인공신경망 기반 기계번역)*의 약어
    * 인공지능 분야 중 하나인 인공신경망을 번역 기술에 적용한 것으로, 인공신경망은 인간이 생각을 하는 최소 단위인 뉴련의 집합체(신경망)를 소프트웨어적으로 구현한 것을 말한다. 즉, 소프트웨어를 인간의 뇌와 유사한 구조로 만들고 번역 능력을 학습시키는 것이다.
    * 인공신경망은 문장의 정보를 가상공간의 특정 지점을 의미하는 벡터(좌표값)로 변환한다. 벡터에는 단어, 구절, 어순 등의 정보가 전부 들어있기 때문에 NMT는 문맥을 이해한 문장 단위의 번역이 가능하다.
    * 파파고의 NMT 기술은 입력 문장을 문장벡터로 변환하는 신경망(encoder)과 문장벡터에서 번역하는 언어의 문장을 생성하는 신경망(decoder)를 대규모의 병렬 코퍼스부터 자동으로 학습한다.
  * N2MT (Naver Neural Machine Translation)
    * NAVER LABS에서 자체 개발한 인공신경망(Artificial Neural Network) 기반 번역 서비스
  * 광학 문자 인식(OCR) 기술

## 음성 분야<br>
### **구글 어시스턴트**<br>
1. 구글 어시스턴트란?<br>
  * 구글에서 제공하는 **인공지능 음성비서** 서비스

  * 대화 기능을 활용해 여러가지 명령을 한번에 요구하는 등 **쌍방향 소통** 가능
  * 목소리로 결제를 승인하는 **보이스 매치(Voice Match)** 기능 테스트 중
  * 자연어 인식 기반 유튜브 검색 기능 제공
<br>

  ![구글 어시스턴트](https://assistant.google.com/static/images/platforms/hero-home.png)

2. 구글 어시스턴트에 사용된 기술<br>

<br>

  ![웨이브넷](https://lh3.googleusercontent.com/Zy5xK_i2F8sNH5tFtRa0SjbLp_CU7QwzS2iB5nf2ijIf_OYm-Q5D0SgoW9SmfbDF97tNEF7CmxaL-o6oLC8sGIrJ5HxWNk79dL1r7Rc=w1440-rw-v1)
<br>

  * **웨이브넷(WaveNet) 기술**
    * WaveNet은 한번에 샘플 하나씩 오디오 신호의 원시 파형을 직접 모델링 한다. WaveNet은 더 자연스럽게 들리는 음성을 낼 수 있을 뿐만 아니라 원시 파형을 사용하면 음악을 포함한 모든 종류의 오디오를 모델링 할 수 있다.
    * WaveNet은 텍스트로부터 음성을 생성해내는 딥러닝 네트워크로서 기존의 방법에 비해 사람의 목소리나 억양 등이 자연스럽게 들리도록 음성을 낼 수 있다. 기존의 *TTS(Text-To-Speech)* 시스템이 짧은 단위의 오디오 파일을 이어붙이는 방식으로 음성을 생성해내는 데에 반해, WaveNet과 같은 *Deep Generative Model*은 거대한 텍스트와 음성 데이터셋을 기반으로 학습해서 문맥에 따른 억양, 발음 등을 반영해 문장 전체에 대한 오디오 파일을 만들어낸다.
    * WaveNet은 현존하는 최고의 TTS 시스템보다 더 자연스러운 음성을 생성하며 인간 수준의 성능과의 차이를 50%이상 줄일 수 있다.

<br>

  ![웨이브넷2](https://lh3.googleusercontent.com/gvNa8TrRd18cG-VWJndWqwIOHXA933mR-ebwwdj-jkZLv1AFpZ3W0jOpwN9UzJnEpvzqCHLvMxDCM5T0XfwV9m0WVlYCf13BSsbLlsA=w1440-rw-v1)
<br>

  * RNN(Recurrent Neural Networks)
    * 구글은 인공지능 신경망인 *RNN(Recurrent Neural Networks)*을 활용해 완전히 새로운 음성 인식 및 언어 이해 모델을 개발해 AI의 몸집을 획기적으로 줄였다.
    * 기존에는 클라우드상에서 100GB 용량의 모델로 돌아가던 AI가 0.5GB 수준 용량의 칩으로 스마트폰 안에 들어갈 수 있다.

## 이미지 분야<br>
### **빅스비 비전**<br>
1. 빅스비 비전이란?<br>
  * 삼성전자가 제공하는 멀티 디바이스를 기반으로 한 개인화 **인공지능 플랫폼** 서비스
  * 렌즈 모드(Lens Mode): 카메라를 실행해 비추기만 하면 피사체와 관련된 정보를 확인할 수 있는 기능
  * 빅스비 비전으로 이미지를 비추면 해당 이미지에 대한 모든 정보를 한 번에 확인하고, 제공받은 이미지 인식 옵션을 활용할 수 있다.
  * GPS와 AR 기술을 바탕으로 사용자가 위치한 장소에 대한 위치 정보, 날씨 정보, 주변 상가 정보 등 다양한 정보를 제공
  * 텍스트 실시간 번역: 카메라에 비친 글씨를 인식해 자동으로 문자를 추출하고 사용자가 설정한 언어로 실시간 번역
<br>

  ![삼성전자 뉴스룸](https://img.kr.news.samsung.com/kr/wp-content/uploads/2018/05/bv-1t-11.png)
<br>

  ![빅스비 비전2](https://img.kr.news.samsung.com/kr/wp-content/uploads/2018/05/bv-6s.png)
<br>

2. 빅스비 비전에 사용된 기술<br>
  * **광학 문자 인식(OCR) 기술**<br>
    * *Optical Character Recognitiond(광학 문자 인식)*의 약어
    * 이미지나 문서의 텍스트를 인식하여 디지털 데이터로 변환시켜주는 자동인식기술
    * 딥러닝 기반 문자 인식 기능 : 개별 문자 이미지를 수집 및 분류하여 문자 데이터베이스를 구축한다. 딥러닝 기반 문자 인식 알고리즘을 이용하여 문자 학습을 수행하고 학습된 정보를 생성 및 적용한다.
    * OCR은 프로세스의 효율화 및 자동화 측면에서 *RPA((Robotic Process Automation)*와 연계했을 때 더욱 효과적으로 사용할 수 있다. RPA는 단순 반복적인 업무 프로세스를 소프트웨어 로봇이 수행하게 하는 업무자동화 기술이다. OCR만으로는 인식한 값들을 저장 및 전송하는 것에 그친다면 RPA와 연계하여 그 다음 처리되는 사람의 업무까지 프로세스 전체를 자동화할 수 있다.

<br>

  ![RPA](https://www.2e.co.kr/news/photo/201810/img_203023_5.jpg)

  * AR 기술
    * *Augmented Reality(증강현실)*의 약어
    * 현실의 이미지나 배경에 가상의 이미지를 추가해 보여주는 기술
    * 증강현실을 구현하기 위해서는 배경을 정확히 인식하는 과정이 필요하다. 카메라 트래킹, 위치정보(GPS), 전자나침반, 중력센서 기술을 통해 주변의 건물이나 조형물을 통해 현재 위치와 방향, 기울어진 정도에 대한 정보를 파악한다. 파악한 정보는 중앙 데이터 서버에 전달되고, 서버는 해당 위치정보와 관련된 부가정보를 앱과 기기로 전달하여 카메라의 화면을 통해 증강현실을 구현한다.

## 자율주행 분야<br>
### **자율주행 자동차**<br>
1. 자율주행 자동차란?<br>
  * 운전자가 조작하지 않아도 주행환경을 인식해 위험을 판단하고 주행경로를 계획해 스스로 운전하는 자동차
  * 감지시스템, 중앙제어장치, 액추에이터 등으로 구성되며 로봇 및 컴퓨터공학, GPS, 정밀센서, 전자제어 등 첨단기술이 필요하다.
<br>

  ![자율주행](https://img.kr.news.samsung.com/kr/wp-content/uploads/2018/04/2.png)
<br>

2. 자율주행 자동차에 사용된 기술<br>
  * **딥러닝(Deep Learning)**
    * 컴퓨터가 스스로 학습할 수 있게 하기 위해 인공 신경망(Artificial Neural Network, ANN)을 기반으로 구축한 기계 학습 기술
    * 시각 인식 지능의 적용
      * 단순한 사물 인식 수준을 넘어 보행자의 움직임, 차량의 진행 방향, 차도와 인도의 구별 등 이미지 내 인식된 사물의 문맥적 의미(Context Awareness)를 이해하고, 이는 인식된 사물에 따라 각기 다른 기능의 차량 제어 기술의 구현으로 이어진다.
    * 학습 지능의 적용: 사람처럼 주행하며 배우기
      *  딥러닝 기반의 인공지능이 장착된 차량을 사람이 운전하면 인공지능이 운전자의 주행 과정 및 다른 차량을 관찰하며 운전하는 방법을 스스로 학습한다. 따라서 인공지능이 주행 데이터를 많이 학습할 수록 자율주행 기능의 완성도가 높아진다.
      * 딥러닝 기반의 자율주행 구현 방식은 인공지능, 특히 딥러닝 관련 역량과 주행 데이터가 핵심적인 역할을 한다.
    * 강화학습(Reinforcement Learning)의 적용: 스스로 규칙을 터득하기
      * 강화학습은 인간의 개입 없이 반복학습을 통해 인공지능이 스스로 목적을 달성하는 과정을 터득해 내는 방법이다. 인간은 인공지능이 달성해야 하는 목적과 보상(reward) 값만 정의해 주면 된다. 인공지능은 수십, 수백만 번의 시행착오를 반복하며 보상 값을 극대화하고 목적을 달성할 수 있는 방법을 스스로 찾아낸다. 특히 주행 데이터 확보의 제약으로 인해 충분한 학습이 어려운 경우-신호등이 없는 교차로, 비보호 좌회전, 우회전, 램프 진입 등-에 유용하게 활용된다.
    * 사람처럼 사고하는 지능의 적용
      * 관계형 추론(Relational Reasoning)이 가능한 인공지능의 구현
      * 지능의 이식(Transferring Intelligence): 유사한 기능을 수행하는 기존 인공지능의 지능을 새로운 인공지능에 이식하여 활용하는 기술
      * 자율주행 기능이 특정 지역에 국한하지 않고 어떤 국가나 지역으로도 쉽게 확장 가능하도록 시스템을 구현하는 것이 가능해진다.
